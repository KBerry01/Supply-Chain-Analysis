```python
# Import dependencies
import pandas as pd
import numpy as np

# suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read in dataset
try:
    df = pd.read_csv('Resources/DataCoSupplyChainDataset.csv', encoding='utf-8')
except UnicodeDecodeError:
    # If a UnicodeDecodeError occurs, try a different encoding
    df = pd.read_csv('Resources/DataCoSupplyChainDataset.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
#Format coulumns to use . notation
df.columns = df.columns.str.replace(" ", "_")
```

```python
df.head()
```

```python
#Drop unwanted columns
cols = ["Product_Image", "Product_Card_Id", "Product_Status", "Product_Description", "Order_Zipcode", "Category_Id", "Customer_Zipcode", "Customer_Email", "Customer_Password"]
df = df.drop(cols, axis = 1)
df.head()
```

```python
df.info()
```

```python
df.Customer_Lname.isna().value_counts()            
```

```python
# Drop exisiting na rows
df = df.dropna()
df.info()
```

```python
#Change data types for Days_for_shipping_(real)
df.Days_for_shipping_real = df["Days_for_shipping_(real)"].astype(int)
df.Days_for_shipping_real.info()
```

```python
df.loc[df['Customer_State'] != 91732 & 95758]
```

```python
# Assign 0 to no_risk and 1 risky
df.Late_delivery_risk = df.Late_delivery_risk.replace(0, "no_risk")
df.Late_delivery_risk = df.Late_delivery_risk.replace( 1, "risk")
df.Late_delivery_risk
```

```python
df.head()
```

```python
#drop '91732', '95758' from customer state for dataframe
# Drop '91732' and '95758' from the Customer_State
df = df[~df['Customer_State'].isin(['91732', '95758'])]
df.info()
```

```python
#Check fpr removal
df[df['Customer_State'].isin(['91732', '95758'])]
```

```python
 # Drop the rows with index 91732 and 95758 for customer state
df['Customer_State'].isin(['91732', '95758']).value_counts()
```

```python
df.to_csv("Data/cleaned_df.csv", index=False)
```
